## index and category of images

```python
#make dictionary for "<category index> : <directory name of the category>"
index_dirname = {}
index_list = []
for ind_name in os.listdir(IMAGE_DIR):
    ind = int(ind_name.split('__')[0])
    index_list.append(ind)
    index_dirname[ind] = ind_name
```

```python
for i in sorted(index_dirname):
    print("{0}, {1}".format(i, index_dirname[i]))
```

## sampling some images 

```python
index_samples = sampling_categories(index_list, n_samples = 20)
```

```python
sampling_dir, image_paths, image_indices = sampling_images(index_samples, n_pics=1, index_dirname=index_dirname)
```
